```python
def Sequenceinterpreter(LsSeq):
    SEQ=fn.Sequence()
    Mark=[0,0,0,0]

    A1=0
    A2=0

    F1=0
    F2=0
    
    P1=0
    P2=0

    for P in LsSeq:
        print(P["descr"])
        t=0
        dt=P["dt"]
        if dt==None or dt==0:
            raise(Exception("Pulses need to have non-zero duration"))
        AUX=[P["aux1"],P["aux2"],P["aux3"],P["aux4"]]
        for m in range(AUX):
            if AUX[m]!=None:
                Mark[m]=AUX[m]
        

        if P["ks_amplitude1"]!=None:
            A1=P["ks_amplitude1"]
        if P["ks_amplitude2"]!=None:
            A2=P["ks_amplitude2"]

        if P["ks_frequency1"]!=None:
            F1=P["ks_frequency1"]
        if P["ks_frequency2"]!=None:
            F2=P["ks_frequency2"]

        if P["ks_phase1"]!=None:
            P1=P["ks_phase1"]
        if P["ks_phase2"]!=None:
            P2=P["ks_phase2"]

        if A1!=0 and A2!=0:
            phase11=P1
            phase21=P2
            phase12,phase22=phase11+np.pi/2,phase21+np.pi/2
            if dt>1e-4:
                print("long duration two-tone pulses can cause unpredictable memory issues for the HDAWG")

            SEQ.addtwotone(dur=dt,phase11=phase11,phase12=phase12,phase21=phase21,phase22=phase22,amp1=A1,amp2=A2,freq1=F1,freq2=F2,phasetime=t,Mark=Mark)
        
        elif A1!=0:
            phase1=P1
            phase2=phase1+np.pi/2
            SEQ.singletone(dur=dt,amp=A1,freq=F1,phase1=phase1,phase2=phase2,phasetime=t,Mark=Mark)
        elif A2!=0:
            phase1=P2
            phase2=phase1+np.pi/2
            SEQ.singletone(dur=dt,amp=A1,freq=F2,phase1=phase1,phase2=phase2,phasetime=t,Mark=Mark)
        
        else:
            SEQ.addwait(dur=dt,Mark=Mark)
        t+=dt
    SEQ.Upload(device,daq)
```

```python
def Psequential_to_nonseq(Ps):
    Pulses_nonseq=[]
    t=0
    for i in len(Ps):
        p={}
        p["t"]=t
        p["tdur"]=Ps[i]["dt"]
        p["freq1"]=Ps[i]["ks_frequency1"]
        p["freq2"]=Ps[i]["ks_frequency2"]
        A1=(Ps["ks_amplitude1"]!=0)
        A2=(Ps["ks_amplitude2"]!=0)
        p["phase1"]=Ps['ks_phase1']
        p["phase2"]=Ps['ks_phase2']
        if A1 and A2:
            p["type"]="Double"
            p["amp2"]=Ps[i]["ks_amplitude2"]
            p["amp1"]=Ps[i]["ks_amplitude1"]
            Pulses_nonseq.append(p)
        elif A1 or A2:
            p["type"]="Single"
            if A2:
                p["amp1"]=Ps[i]["ks_amplitude2"]
                p["amp2"]=Ps[i]["ks_amplitude1"]
            else:
                p["amp2"]=Ps[i]["ks_amplitude2"]
                p["amp1"]=Ps[i]["ks_amplitude1"]
            Pulses_nonseq.append(p)
        
        t+=p["tdur"]
    return Pulses_nonseq
        
```
